# Regression using gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data points (x, y)
X = np.array([0, 1, 2])
Y = np.array([0, 2, 8])

# Initialize regression coefficients
# y = ax_1+bx_0  (x_0=1)

a = -1  # slope
b = 2   # intercept

# Learning rate
alpha = 0.5

# Number of iterations
iterations = 50


# Store the regression lines for plotting
plt.figure(figsize=(8,6))
plt.scatter(X, Y, color='blue', label='Data points')


# Gradient descent iterations
for iteration in range(iterations):
    # Calculate predictions
    Y_pred = a * X + b

    # Plot the current line
    plt.plot(X, Y_pred, label=f'Iteration {iteration+1}', alpha=0.3)

    # Compute gradients
    grad_a = -np.dot(X, (Y - Y_pred)) / len(X)
    grad_b = -np.sum(Y - Y_pred) / len(X)
    print(grad_a,grad_b)

    # Update coefficients
    a = a - alpha * grad_a
    b = b - alpha * grad_b
    print(a,b)

# Final plot settings
plt.title('Linear Regression using Gradient Descent')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.ylim([-1,10])
plt.show()
```
