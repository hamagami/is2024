# 06 Decision Tree Regression by sklearn

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Example dataset with features for regression
data = {
    'Temperature': [30, 22, 25, 28, 35, 40, 20, 18, 15, 25],
    'Humidity': [70, 65, 80, 75, 60, 50, 85, 90, 95, 85],
    'Energy_Consumption': [200, 150, 160, 180, 210, 250, 140, 130, 120, 155]
}

df = pd.DataFrame(data)

# Features and target variable
X = df[['Temperature', 'Humidity']]
y = df['Energy_Consumption']
```

```python
# Create the regression tree model
regressor = DecisionTreeRegressor(max_depth=3, random_state=42)
regressor.fit(X, y)

# Visualize the regression tree
plt.figure(figsize=(12, 8))
plot_tree(regressor, feature_names=['Temperature', 'Humidity'], filled=True, precision=1)
plt.title("Regression Tree for Energy Consumption Prediction")
plt.show()

# Predicting on new data points
new_data = np.array([[32, 72], [20, 80], [38, 60]])
predictions = regressor.predict(new_data)
print("Predicted Energy Consumption for new data points:", predictions)
```
