# Polynomial regression

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data points that follow a quadratic curve with some noise
np.random.seed(42)
X = np.linspace(-10, 10, 100)
y_true = 0.5 * X**2 + 2 * X + 5  # True quadratic function
noise = np.random.normal(0, 10, size=X.shape)  # Adding some noise to the data
y_noisy = y_true + noise

# Initialize parameters for quadratic model: y = a*x^2 + b*x + c
a = 0.0
b = 0.0
c = 0.0

# Learning rate
alpha = 0.0001

# Number of iterations (epochs)
epochs = 2000

# Number of data points
n = len(X)

# Perform SGD to fit the quadratic model
for epoch in range(epochs):
    for i in range(n):
        x_i = X[i]
        y_i = y_noisy[i]

        # Predicted value
        y_pred = a * x_i**2 + b * x_i + c

        # Gradients for a, b, and c
        grad_a = -2 * (y_i - y_pred) * (x_i**2)
        grad_b = -2 * (y_i - y_pred) * x_i
        grad_c = -2 * (y_i - y_pred)

        # Update parameters
        a = a - alpha * grad_a
        b = b - alpha * grad_b
        c = c - alpha * grad_c

# Predicted values using the learned parameters
y_pred_final = a * X**2 + b * X + c

# Plot the noisy data, true quadratic function, and the predicted quadratic function
plt.figure(figsize=(8,6))
plt.scatter(X, y_noisy, color='blue', label='Noisy Data')
plt.plot(X, y_true, color='red', label='True Quadratic Function')
plt.plot(X, y_pred_final, color='green', label='Predicted Quadratic Function (SGD)')
plt.title('Quadratic Regression with SGD')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Print the learned parameters
print(f"Learned parameters: a = {a}, b = {b}, c = {c}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate complex data combining sine, exponential, and noise
np.random.seed(42)
X = np.linspace(0, 10, 200)
y_true = 5 * np.sin(2 * X) + 0.5 * np.exp(0.3 * X)  # True function (sin + exp)
noise = np.random.normal(0, 1, size=X.shape)  # Adding some noise
y_noisy = y_true + noise  # Noisy data

# Given values for B and D
B = 2  # Given fixed value for the sine component
D = 0.3  # Given fixed value for the exponential component

# Initialize parameters for the complex model: y = A*sin(B*x) + C*exp(D*x) + E
A = 1.0
C = 1.0
E = 1.0

# Learning rate
alpha = 0.0001

# Number of iterations (epochs)
epochs = 1000

# Number of data points
n = len(X)

# Perform SGD to fit the complex model
for epoch in range(epochs):
    for i in range(n):
        x_i = X[i]
        y_i = y_noisy[i]

        # Predicted value
        y_pred = A * np.sin(B * x_i) + C * np.exp(D * x_i) + E

        # Gradients for A, C, and E (B and D are fixed)
        grad_A = -2 * (y_i - y_pred) * np.sin(B * x_i)
        grad_C = -2 * (y_i - y_pred) * np.exp(D * x_i)
        grad_E = -2 * (y_i - y_pred)

        # Update parameters
        A = A - alpha * grad_A
        C = C - alpha * grad_C
        E = E - alpha * grad_E

# Predicted values using the learned parameters
y_pred_final = A * np.sin(B * X) + C * np.exp(D * X) + E

# Plot the noisy data, true function, and the predicted function
plt.figure(figsize=(10,6))
plt.scatter(X, y_noisy, color='blue', label='Noisy Data', s=10)
plt.plot(X, y_true, color='red', label='True Function (sin + exp)', linewidth=2)
plt.plot(X, y_pred_final, color='green', label='Predicted Function (SGD)', linewidth=2)
plt.title('Fitting a Sine + Exponential Model Using SGD (Fixed B and D)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Print the learned parameters
print(f"Learned parameters: A = {A}, C = {C}, E = {E}")
```
