# k-means sklearn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Step 1: Generate synthetic data
def generate_data():
    """
    Generate synthetic data with 3 clusters for demonstration.
    """
    np.random.seed(42)
    data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
    return data

# Step 2: Perform k-means clustering using scikit-learn
def kmeans_clustering(data, n_clusters=3):
    """
    Perform k-means clustering using Scikit-learn's KMeans class.
    - data: Input data (numpy.ndarray)
    - n_clusters: Number of clusters
    """
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    return labels, centroids

# Step 3: Visualize the clustering results
def visualize_clusters(data, labels, centroids):
    """
    Visualize the data points and centroids after clustering.
    - data: Input data
    - labels: Cluster labels for each data point
    - centroids: Coordinates of the final centroids
    """
    plt.figure(figsize=(8, 6))
    # Plot data points color-coded by their cluster
    for i in range(np.max(labels) + 1):
        plt.scatter(data[labels == i, 0], data[labels == i, 1], label=f'Cluster {i + 1}')
    # Plot centroids
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
    plt.legend()
    plt.title("K-means Clustering (Scikit-learn)")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

```

```python
# Main execution
if __name__ == "__main__":
    # Generate data
    data = generate_data()
    # Perform k-means clustering
    labels, centroids = kmeans_clustering(data, n_clusters=3)
    # Visualize results
    visualize_clusters(data, labels, centroids)
```
