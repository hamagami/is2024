#  Hierarchical clustering using Single-linkage

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.datasets import make_blobs
```

```python

# Generate synthetic data
def generate_data():
    """
    Generate synthetic data for clustering.
    """
    np.random.seed(42)
    X, _ = make_blobs(n_samples=50, centers=3, cluster_std=1.5, random_state=42)
    return X

# Perform hierarchical clustering using Single-linkage
def hierarchical_clustering(X):
    """
    Perform hierarchical clustering using Single-linkage and plot the dendrogram.
    - X: Input data (numpy.ndarray)
    """
    # Step 1: Compute the linkage matrix using 'single' linkage
    linkage_matrix = linkage(X, method='single')

    # Step 2: Plot the dendrogram
    plt.figure(figsize=(10, 7))
    dendrogram(linkage_matrix, labels=np.arange(1, X.shape[0] + 1))
    plt.title("Hierarchical Clustering Dendrogram (Single-linkage)")
    plt.xlabel("Sample Index")
    plt.ylabel("Distance")
    plt.show()

# Visualize the clustered data points
def visualize_data(X):
    """
    Scatter plot of the original data.
    - X: Input data
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], s=50, color='blue', edgecolors='k')
    plt.title("Input Data for Hierarchical Clustering")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()


```

```python
# Main execution
if __name__ == "__main__":
    # Step 1: Generate synthetic data
    data = generate_data()

    # Step 2: Visualize input data
    visualize_data(data)

    # Step 3: Perform hierarchical clustering with Single-linkage
    hierarchical_clustering(data)
```
