# 05_SVM_regression_with_scikit-learn

SVM is not only for classification problems; it can also be used for regression.
By using scikit-learn, nonlinear SVM regression can be implemented easily.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
# Generate synthetic data for regression (non-linear relationship)
np.random.seed(0)
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel() + 0.1 * np.random.randn(40)

# Fit regression model using SVM with RBF kernel
svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1)
svr_rbf.fit(X, y)

# Predict and evaluate
X_test = np.linspace(0, 5, 100).reshape(-1, 1)
y_pred = svr_rbf.predict(X_test)
mse = mean_squared_error(y, svr_rbf.predict(X))
```

```python
# Plot results
plt.scatter(X, y, color='darkorange', label='Data')
plt.plot(X_test, y_pred, color='navy', lw=2, label='SVR model (RBF kernel)')
plt.xlabel("X")
plt.ylabel("y")
plt.title("SVM Regression with RBF Kernel")
plt.legend()
plt.show()

print("Mean Squared Error on Training Data:", mse)
```
